```python
import sympy as sp

# Define the symbol and function
X = sp.Symbol('x')
function = 'x**2 - 6'  # Example function
func = sp.sympify(function)  # Convert the string to a sympy expression
func_deriv = sp.diff(func, X)  # Compute the first derivative symbolically
func_d2 = sp.diff(func_deriv, X)  # Compute the second derivative symbolically

# Define the function and its derivatives
def f(x_val):
    return func.subs(X, x_val).evalf()  # Substitute x_val into the function and evaluate

def df(x_val):
    return func_deriv.subs(X, x_val).evalf()  # Substitute x_val into the first derivative and evaluate

def d2f(x_val):
    return func_d2.subs(X, x_val).evalf()  # Substitute x_val into the second derivative and evaluate

# Define Halley's method
def halley_method():
    p0 = 1  # Initial guess
    N = 100  # Maximum number of iterations
    error = 1e-8  # Precision required

    i = 1
    while i <= N:
        fx = f(p0)
        dfx = df(p0)
        d2fx = d2f(p0)
        
        denominator = 2 * (dfx ** 2) - fx * d2fx
        
        if denominator == 0:
            print("Denominator became zero. Halley's method failed.")
            return None
        
        g_x = p0 - (2 * fx * dfx) / denominator
        
        print(f"Iteration {i}: x = {g_x:.10f}")
        
        if abs(g_x - p0) < error:
            print(f"Solution found: x = {g_x:.10f} after {i} iterations.")
            return g_x
        
        p0 = g_x
        i += 1

    print(f"Solution did not converge within {N} iterations.")
    return None

# Call Halley's method
halley_method()
```

```python
#Jacobi method
import numpy as np

def jacobi_method(A, b, x0, tol=1e-10, max_iter=100):
    """
    Solve the system of linear equations Ax = b using the Jacobi method.
    
    Parameters:
    A : 2D numpy array
        Coefficient matrix.
    b : 1D numpy array
        Right-hand side vector.
    x0 : 1D numpy array
        Initial guess for the solution.
    tol : float, optional
        Tolerance for convergence (default is 1e-10).
    max_iter : int, optional
        Maximum number of iterations (default is 100).
        
    Returns:
    x : 1D numpy array
        Approximate solution to the system.
    """
    n = len(b)
    x = x0.copy()
    
    for k in range(max_iter):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            print(f"Converged after {k+1} iterations")
            return x_new
        
        x = x_new

    print("Did not converge within the maximum number of iterations")
    return x

# Example usage
A = np.array([[4, -1, 0, 0],
              [-1, 4, -1, 0],
              [0, -1, 4, -1],
              [0, 0, -1, 3]], dtype=float)

b = np.array([15, 10, 10, 10], dtype=float)
x0 = np.zeros_like(b)

solution = jacobi_method(A, b, x0)
print("Solution:", solution)
```

```python
# gauss_seidel 
import numpy as np

def gauss_seidel_method(A, b, x0, tol=1e-10, max_iter=100):
    """
    Solve the system of linear equations Ax = b using the Gauss-Seidel method.
    
    Parameters:
    A : 2D numpy array
        Coefficient matrix.
    b : 1D numpy array
        Right-hand side vector.
    x0 : 1D numpy array
        Initial guess for the solution.
    tol : float, optional
        Tolerance for convergence (default is 1e-10).
    max_iter : int, optional
        Maximum number of iterations (default is 100).
        
    Returns:
    x : 1D numpy array
        Approximate solution to the system.
    """
    n = len(b)
    x = x0.copy()
    
    for k in range(max_iter):
        x_new = x.copy()
        for i in range(n):
            s = sum(A[i][j] * x_new[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            print(f"Converged after {k+1} iterations")
            return x_new
        
        x = x_new

    print("Did not converge within the maximum number of iterations")
    return x

# Example usage
A = np.array([[4, -1, 0, 0],
              [-1, 4, -1, 0],
              [0, -1, 4, -1],
              [0, 0, -1, 3]], dtype=float)

b = np.array([15, 10, 10, 10], dtype=float)
x0 = np.zeros_like(b)

solution = gauss_seidel_method(A, b, x0)
print("Solution:", solution)
```

```python
# interpolation
#Non equal interval data (Lagrange)
#Lagrange method
def lagrange_interpolation(x_data, y_data, x):
    n = len(x_data)
    result = 0
    
    for i in range(n):
        term = y_data[i]
        for j in range(n):
            if j != i:
                term *= (x - x_data[j]) / (x_data[i] - x_data[j])
        result += term
    
    return result

x_data = [1, 2, 3, 4]
y_data = [1, 4, 9, 16]
x = 2.5

result = lagrange_interpolation(x_data, y_data, x)
print(f"The value of the Lagrange interpolation polynomial at x = {x} is: {result:.2f}")
```

```python
# interpolation
#equal interval data (Newton forward)
import numpy as np

def newton_forward_interpolation(x, y, xi):
    """
    Perform Newton Forward Interpolation for equally spaced data points.
    
    Parameters:
    x : 1D numpy array
        Array of x values (equally spaced).
    y : 1D numpy array
        Array of y values corresponding to x values.
    xi : float
        The x value at which to interpolate.
        
    Returns:
    yi : float
        Interpolated y value at xi.
    """
    n = len(x)
    h = x[1] - x[0]  # Assuming equally spaced data points
    
    # Create a forward difference table
    forward_diff = np.zeros((n, n))
    forward_diff[:, 0] = y
    
    for j in range(1, n):
        for i in range(n - j):
            forward_diff[i, j] = forward_diff[i + 1, j - 1] - forward_diff[i, j - 1]
    
    # Print the forward difference table
    print("Forward Difference Table:")
    for i in range(n):
        for j in range(n - i):
            print(f"{forward_diff[i, j]:.2f}", end="\t")
        print()
    
    # Compute the interpolation value at xi
    yi = y[0]
    u = (xi - x[0]) / h
    u_term = 1
    
    for i in range(1, n):
        u_term *= (u - i + 1) / i
        yi += u_term * forward_diff[0, i]
    
    return yi

# Example usage
x = np.array([0, 1, 2, 3, 4], dtype=float)
y = np.array([1, 2, 4, 8, 16], dtype=float)
xi = 2.5

yi = newton_forward_interpolation(x, y, xi)
print(f"\nInterpolated value at x = {xi} is y = {yi:.2f}")


```

```python
# interpolation
#equal interval data (Newton Bacward)
import numpy as np

def newton_backward_interpolation(x, y, xi):
    """
    Perform Newton Backward Interpolation for equally spaced data points.
    
    Parameters:
    x : 1D numpy array
        Array of x values (equally spaced).
    y : 1D numpy array
        Array of y values corresponding to x values.
    xi : float
        The x value at which to interpolate.
        
    Returns:
    yi : float
        Interpolated y value at xi.
    """
    n = len(x)
    h = x[1] - x[0]  # Assuming equally spaced data points
    
    # Create a backward difference table
    backward_diff = np.zeros((n, n))
    backward_diff[:, 0] = y
    
    for j in range(1, n):
        for i in range(n - 1, j - 1, -1):
            backward_diff[i, j] = backward_diff[i, j - 1] - backward_diff[i - 1, j - 1]
    
    # Print the backward difference table
    print("Backward Difference Table:")
    for i in range(n):
        for j in range(i + 1):
            print(f"{backward_diff[i, j]:.2f}", end="\t")
        print()
    
    # Compute the interpolation value at xi
    yi = y[-1]
    u = (xi - x[-1]) / h
    u_term = 1
    
    for i in range(1, n):
        u_term *= (u + i - 1) / i
        yi += u_term * backward_diff[-1, i]
    
    return yi

# Example usage
x = np.array([0, 1, 2, 3, 4], dtype=float)
y = np.array([1, 2, 4, 8, 16], dtype=float)
xi = 2.5

yi = newton_backward_interpolation(x, y, xi)
print(f"\nInterpolated value at x = {xi} is y = {yi:.2f}")

```

```python
# diffrentiation 2 point
def differentiate(x, y):
    """
    Perform differentiation using different methods based on the position of points in the table.
    Forward difference for the first point, backward difference for the last point,
    and central difference for the middle points.
    
    Parameters:
    x : 1D numpy array
        Array of x values (equally spaced).
    y : 1D numpy array
        Array of y values corresponding to x values.
        
    Returns:
    dy_dx : 1D numpy array
        Array of approximate derivatives corresponding to x values.
    """
    n = len(x)
    h = x[1] - x[0]  # Assuming equally spaced data points
    
    # Compute differentiation using forward difference for the first point
    dy_dx = np.zeros_like(y)
    dy_dx[0] = (y[1] - y[0]) / h
    
    # Compute differentiation using central difference for the middle points
    for i in range(1, n - 1):
        dy_dx[i] = (y[i + 1] - y[i - 1]) / (2 * h)
    
    # Compute differentiation using backward difference for the last point
    dy_dx[-1] = (y[-1] - y[-2]) / h
    
    return dy_dx

# Example usage
import numpy as np

# Sample data
x = np.array([0, 1, 2, 3, 4], dtype=float)
y = np.array([1, 2, 4, 8, 16], dtype=float)

# Compute differentiation using different methods
dy_dx = differentiate(x, y)

# Print the result
print("x\t y\t dy_dx")
for i in range(len(x)):
    print(f"{x[i]:.2f}\t {y[i]:.2f}\t {dy_dx[i]:.2f}")

```

```python
def differentiate_3_point(x, y):
    """
    Perform 3-point differentiation using different methods based on the position of points in the table.
    Forward difference for the first point, backward difference for the last point,
    and central difference for the middle points.
    
    Parameters:
    x : 1D numpy array
        Array of x values (equally spaced).
    y : 1D numpy array
        Array of y values corresponding to x values.
        
    Returns:
    dy_dx : 1D numpy array
        Array of approximate derivatives corresponding to x values.
    """
    n = len(x)
    h = x[1] - x[0]  # Assuming equally spaced data points
    
    # Compute differentiation using forward difference for the first point
    dy_dx = np.zeros_like(y)
    dy_dx[0] = (-3*y[0] + 4*y[1] - y[2]) / (2 * h)
    
    # Compute differentiation using central difference for the middle points
    for i in range(1, n - 1):
        dy_dx[i] = (y[i + 1] - y[i - 1]) / (2 * h)
    
    # Compute differentiation using backward difference for the last point
    dy_dx[-1] = (3*y[-1] - 4*y[-2] + y[-3]) / (2 * h)
    
    return dy_dx

# Example usage
import numpy as np

# Sample data
x = np.array([0, 1, 2, 3, 4], dtype=float)
y = np.array([1, 2, 4, 8, 16], dtype=float)

# Compute differentiation using different methods
dy_dx = differentiate_3_point(x, y)

# Print the result
print("x\t y\t dy_dx")
for i in range(len(x)):
    print(f"{x[i]:.2f}\t {y[i]:.2f}\t {dy_dx[i]:.2f}")
```

```python
# integration
# simpson
def simpsons_rule(f, a, b, n):
    """
    Approximate the definite integral of a function using Simpson's rule.
    
    Parameters:
    f : function
        The function to be integrated.
    a : float
        The lower limit of integration.
    b : float
        The upper limit of integration.
    n : int
        The number of intervals. Must be even.
        
    Returns:
    integral : float
        Approximated value of the definite integral.
    """
    # Check if the number of intervals is even
    if n % 2 != 0:
        raise ValueError("Number of intervals must be even.")
    
    # Calculate the width of each interval
    h = (b - a) / n
    
    # Generate equally spaced points from a to b
    x = np.linspace(a, b, n+1)
    
    # Evaluate the function at these points
    y = f(x)
    
    # Apply Simpson's rule formula to compute the approximate integral
    integral = h / 3 * (y[0] + 4 * np.sum(y[1:n:2]) + 2 * np.sum(y[2:n-1:2]) + y[-1])
    
    return integral

# Example usage
import numpy as np

# Define the function to be integrated
def f(x):
    return x**2

# Define the limits of integration and the number of intervals
a = 0
b = 2
n = 10  # Must be even

# Compute the definite integral using Simpson's rule
integral = simpsons_rule(f, a, b, n)
print("Approximated value of the definite integral:", integral)
```

```python
# integration
# Trapezoidale
def trapezoidal_rule(f, a, b, n):
    """
    Approximate the definite integral of a function using the trapezoidal rule.
    
    Parameters:
    f : function
        The function to be integrated.
    a : float
        The lower limit of integration.
    b : float
        The upper limit of integration.
    n : int
        The number of intervals.
        
    Returns:
    integral : float
        Approximated value of the definite integral.
    """
    h = (b - a) / n  # Calculate the width of each interval
    x = np.linspace(a, b, n+1)  # Generate equally spaced points from a to b
    y = f(x)  # Evaluate the function at these points
    integral = h / 2 * (y[0] + 2 * np.sum(y[1:n]) + y[-1])  # Apply the trapezoidal rule formula
    return integral

# Example usage
import numpy as np

# Define the function to be integrated
def f(x):
    return x**2

# Define the limits of integration and the number of intervals
a = 0
b = 2
n = 10

# Compute the definite integral using the trapezoidal rule
integral = trapezoidal_rule(f, a, b, n)
print("Approximated value of the definite integral:", integral)
```

```python
# difrential equation
#Solving IVP using euilar method
#Solving IVP using euilar method
import numpy as np
import matplotlib.pyplot as plt

# Solver function
def solver_IVP(f, tspan, y0, h, solver):
    t = np.arange(tspan[0], tspan[1] + h, h)
    y = np.zeros(len(t))
    y[0] = y0
    
    for n in range(len(t) - 1):
        y[n + 1] = solver(f, t[n], y[n], h)
    
    return t, y

# Euler method
def euler(f, tn, yn, h):
    return yn + h * f(tn, yn)

# Define IVP
def f(t, y):
    return t * y

tspan = [0, 1]
y0 = 1
h = 0.2

# Solve IVP
t, y = solver_IVP(f, tspan, y0, h, euler)

# Print table of solutions
print(" t |    y\n-------------------")
for n in range(len(t)):
    print(f"{t[n]:0.1f} | {y[n]:10.6f}")

# Plot solutions
fig, ax = plt.subplots()
plt.plot(t, y, "b-o")  # corrected "b-o" for blue line with circles
plt.xlabel("$t$")
plt.ylabel("$y$")
plt.show()
```

```python
# difrential equation
#Solving IVP using Runge kutta
# Constants
import math
import matplotlib.pyplot as plt
T_0 = 0
Y_0 = 1

T_FINAL = 5
DT = 0.5
DT_EXACT = 0.01


# Functions
def f(t, y):
    return math.sin(t)**2 * y

def euler_step(t_i, y_i, dt, f):
    return y_i + dt*f(t_i, y_i)

def rk4_step(t_i, y_i, dt, f):
    k1 = f(t_i, y_i)
    k2 = f(t_i + dt/2, y_i + k1*dt/2)
    k3 = f(t_i + dt/2, y_i + k2*dt/2)
    k4 = f(t_i + dt, y_i + k3*dt)
    return y_i + dt/6*(k1 + 2*k2 + 2*k3 + k4)

def y_exact(t_0, y_0, t):
    exp_arg = 1/2 * ( (t-t_0) - (math.sin(t)*math.cos(t) - math.sin(t_0)*math.cos(t_0)) )
    return y_0*math.exp(exp_arg)


# Computing exact result
t = T_0

ys_exact = []
ts_exact = []

while t < T_FINAL:
    ts_exact.append(t)
    ys_exact.append(y_exact(T_0, Y_0, t))
    t += DT_EXACT


# Computing approximate results
ts = [T_0]
ys = [Y_0]
ys_euler = [Y_0]

y = Y_0
y_euler = Y_0
t = T_0

while t < T_FINAL:
    # Solving with Runge-Kutta
    y = rk4_step(t, y, DT, f)

    # Solving with Euler
    y_euler = euler_step(t, y_euler, DT, f)

    # Increasing t
    t += DT

    # Appending results
    ts.append(t)
    ys.append(y)
    ys_euler.append(y_euler)


# Plotting
plt.plot(ts, ys, color='red', marker='o', linewidth=0.0, label='RK4')
plt.plot(ts, ys_euler, color='green', marker='o', linewidth=0.0, label='Euler')
plt.plot(ts_exact, ys_exact, color='blue', label='Exact')
plt.legend()
plt.show()
```
